```python
import sqlite3
```

```python
conn = sqlite3.connect("hostel.db")
cursor = conn.cursor()

print("Database connected!")
```

```python
# STUDENT TABLE
cursor.execute("""
CREATE TABLE IF NOT EXISTS student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    room_id INTEGER
)
""")

# ROOM TABLE
cursor.execute("""
CREATE TABLE IF NOT EXISTS room (
    room_id INTEGER PRIMARY KEY,
    room_type TEXT,
    capacity INTEGER
)
""")

# FEES TABLE
cursor.execute("""
CREATE TABLE IF NOT EXISTS fees (
    fee_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    amount INTEGER,
    status TEXT
)
""")

# COMPLAINT TABLE
cursor.execute("""
CREATE TABLE IF NOT EXISTS complaint (
    complaint_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    issue TEXT
)
""")

conn.commit()

print("All tables created successfully!")
```

```python
# insert room
cursor.execute("INSERT INTO room VALUES (101, 'AC', 2)")
cursor.execute("INSERT INTO room VALUES (102, 'Non-AC', 3)")

# insert student
cursor.execute("INSERT INTO student VALUES (1, 'SK', 'IT', 101)")
cursor.execute("INSERT INTO student VALUES (2, 'Arun', 'CSE', 102)")

# insert fees
cursor.execute("INSERT INTO fees (student_id, amount, status) VALUES (1, 50000, 'Paid')")
cursor.execute("INSERT INTO fees (student_id, amount, status) VALUES (2, 50000, 'Pending')")

conn.commit()

print("Sample data inserted!")
```

```python
def add_student():
    id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    dept = input("Enter Dept: ")
    room = int(input("Enter Room: "))

    cursor.execute("INSERT INTO student VALUES (?, ?, ?, ?)", (id, name, dept, room))
    conn.commit()
    print("Student Added!")


def view_students():
    cursor.execute("SELECT * FROM student")
    for row in cursor.fetchall():
        print(row)


def update_student():
    id = int(input("Enter Student ID: "))
    room = int(input("Enter New Room ID: "))

    cursor.execute("UPDATE student SET room_id=? WHERE student_id=?", (room, id))
    conn.commit()
    print("Updated!")


def delete_student():
    id = int(input("Enter Student ID: "))

    cursor.execute("DELETE FROM student WHERE student_id=?", (id,))
    conn.commit()
    print("Deleted!")


# 🔁 SAME while loop style
while True:
    print("\n1.Add\n2.View\n3.Update\n4.Delete\n5.Exit")
    ch = int(input("Enter choice: "))

    if ch == 1:
        add_student()

    elif ch == 2:
        view_students()

    elif ch == 3:
        update_student()

    elif ch == 4:
        delete_student()

    elif ch == 5:
        print("Exiting...")
        break

    else:
        print("Invalid choice!")
```
